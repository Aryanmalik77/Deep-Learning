```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
from keras.layers import Dropout
from sklearn.datasets import make_moons
from mlxtend.plotting import plot_decision_regions
from keras.optimizers import Adam
```

```python
x,y = make_moons(n_samples = 100, noise = 0.5, random_state = 42)
```

```python
plt.scatter(x[:,0],x[:,1], c= y)
```

```python
model1 = Sequential()
model1.add(Dense(10,activation = "relu", input_dim = 2))
model1.add(Dense(10, activation = "relu"))
model1.add(Dense(1, activation = "sigmoid"))
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
history = model1.fit(x,y , epochs = 200)
```

```python
plot_decision_regions(x,y,clf = model1,legend =2)

plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
# fixing of the overfitting
```

```python
model2 = Sequential()
model2.add(Dense(10,activation = "relu", input_dim = 2,kernel_regularizer=tf.keras.regularizers.l1(0.001)))
model2.add(Dense(10,activation = "relu", kernel_regularizer=tf.keras.regularizers.l1(0.001)))
model2.add(Dense(1,activation = "sigmoid", ))
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
history = model1.fit(x,y , epochs = 200)

```

```python
plot_decision_regions(x,y,clf = model2,legend =2)

plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```
